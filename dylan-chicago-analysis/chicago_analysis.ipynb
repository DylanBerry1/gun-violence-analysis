```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('chicago_violence.csv')
```

```python
df.columns
```

```python
df[df['GUNSHOT_INJURY_I'] == 'YES'].groupby('TIME_PERIOD').count().reset_index().sort_values(by='TIME_PERIOD', ascending=True).set_index('TIME_PERIOD')[['TIME_PERIOD_START']].rename(columns={'TIME_PERIOD_START':'count'}).plot(kind='line')
```

```python
df[df['GUNSHOT_INJURY_I'] == 'YES'].groupby('SEX').count()['TIME_PERIOD'].plot(kind='bar')
```
